# Ejemplo de recursividad

```python
import tqdm
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
def factorial(n):
  if n==0 or n==1: return 1
  else: return n*factorial(n-1)


n=10

start_time = time.time()  # Guardamos el tiempo antes de ejecutar
resultado = factorial(n)  # Calculamos el factorial
end_time = time.time()  # Guardamos el tiempo después de ejecutar

# Mostramos el resultado y el tiempo de ejecución
print(f"El factorial de {n} es {resultado}")
print(f"Tiempo de ejecución: {end_time - start_time} segundos")
```

```python
def faci(n):
  if n==0: return 1

  i=1
  f=1

  while i<n:
    print(i)
    i+=1
    f*=i
  return f
```

```python
n=10

start_time = time.time()  # Guardamos el tiempo antes de ejecutar
resultado = faci(n)  # Calculamos el factorial
end_time = time.time()  # Guardamos el tiempo después de ejecutar

# Mostramos el resultado y el tiempo de ejecución
print(f"El factorial de {n} es {resultado}")
print(f"Tiempo de ejecución: {end_time - start_time} segundos")
```

# Método de diferencias divididas

3.3 Burden

```python
def D(i, k , Xn, Yn):
  # si k+1>N
  if i+k>=len(Xn):
    return 0
  elif k==0:
    return Yn[i]
  else:
    return (D(i+1,k-1,Xn,Yn)-D(i,k-1,Xn,Yn))/(Xn[i+k]-Xn[i])
```

```python
n=8
xi = np.linspace(0,2*np.pi,n)
yi = np.sin(xi)
plt.plot(xi,yi, '+')

espacio_polado = np.linspace(0,2*np.pi,1000)

for x in espacio_polado:
  pn = 0
  for k in range(1,n):
    lk1 = 1
    for i in range(0,k):
      lk1 *= x - xi[i]
    dk = D(0,k,xi,yi)
    pn += yi[0]+ dk*(lk1)
  plt.plot(x,pn,'*')
```
